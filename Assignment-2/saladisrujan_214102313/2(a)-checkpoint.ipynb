```python
#importing required libraries 
import numpy as np
import matplotlib.pyplot as plt
x1_vals=np.linspace(-10,10,100)
x2_vals=np.linspace(-10,10,100)
x1,x2=np.meshgrid(x1_vals,x2_vals) #to form a mesh grid on x1,x2 plane

def eigen(a,b,c,d):  #defining a function to get eigen values and eigen vectors
    m=np.array([[a,b],[c,d]])
    evals,evect=np.linalg.eig(m)
    print(f'the eigen values and eigen vectors are respectively:{evals} , {evect}')
```

```python
def func_y(x1,x2):
    return 0.5*x1**2+0.5*x2**2 #quadratic form of 1st matrix

y=func_y(x1,x2)
ax=plt.axes(projection='3d') #to get a 3d axis
ax.plot_surface(x1,x2,y)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y');

eigen(0.5,0,0,0.5)
```
