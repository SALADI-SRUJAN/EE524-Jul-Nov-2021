```python
import matplotlib.pyplot as plt
import numpy as np
def covariancy(a,b,c,d):  ##funtion to check for covariancy
    m=np.array([[a,b],[c,d]])
    p=np.linalg.eigvals(m)
    if b==c and np.all(p>=0):
        return print('Given matrix is a co-variance matrix')
    else:
        return print('not a co-variance matrix')  
x1_vals=np.linspace(-10,10,100)
x2_vals=np.linspace(-10,10,100)
x1,x2=np.meshgrid(x1_vals,x2_vals) #to get a meshgrid on x1,x2 plane
```

```python
covariancy(-1,2,3,4)
y=-1*x1**2+4*x2**2+5*x1*x2
ax=plt.axes(projection='3d')
ax.plot_surface(x1,x2,y)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y');
```

```python
covariancy(2,4,6,8)
y=2*x1**2+8*x2**2+10*x1*x2
ax=plt.axes(projection='3d')
ax.plot_surface(x1,x2,y)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y');
```
