```python
import matplotlib.pyplot as plt
import numpy as np
x_vals=np.linspace(-10,10,100)
y_vals=np.linspace(-10,10,100)
x,y=np.meshgrid(x_vals,y_vals)
# For the given hessian matrix the possible quadratic form is (x1**2+x2**2+x1*x2) 
```

```python
h=([2,1],[1,2])
# to determine covexity/concavity from the hessian matrix
if np.all(np.linalg.eigvals(h))>=0:
    print('strictly convex')
else:
    print('strictly concave')

z=x**2+y**2+x*y
ax=plt.axes(projection='3d')
ax.plot_surface(x,y,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
