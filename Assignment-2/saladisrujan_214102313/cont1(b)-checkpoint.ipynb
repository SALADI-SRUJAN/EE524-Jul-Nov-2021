```python
#The given distribution follows a chi-square distribution with two degrees of freedom i.e. k=2
#chi-square is defined only for zero mean and unit variance,for a different variance its a scaling factor

#importing required libraries
import numpy as np
import math
import matplotlib.pyplot as plt
x=np.linspace(0,10)
# defining a function for chi square pdf
def chisquare_pdf(k):
    y=np.piecewise(x,[(x>=0)&(x<=10)],[lambda x:(x**((k/2)-1)*(2.718**(-x/2)))/(2**(k/2)*math.gamma(k/2)*0.25)])# multiplied by 0.25 as scalling factor
    return y
power=chisquare_pdf(2)
k=2
plt.plot(x,power,label='power')
mean=k
median=k*(1-(2/9*k))**3
mode=max(k-2,0)
var=2*k
plt.plot(mean,0,"s",label='mean')
plt.plot(median,0,"o",label='median')
plt.plot(mode,0,"d",label='mode')
print(f'the variance is :{var}')
plt.title('power')
plt.xlabel('x')
plt.legend()
plt.show()
```
