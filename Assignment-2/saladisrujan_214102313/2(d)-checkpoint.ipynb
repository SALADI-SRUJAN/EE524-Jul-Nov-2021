```python
import matplotlib.pyplot as plt
import numpy as np
x_vals=np.linspace(-10,10,100)
y_vals=np.linspace(-10,10,100)
x,y=np.meshgrid(x_vals,y_vals)
# To comment about maxima/minima
def max_min(a,b,c,d):
    if ((a*d)-(b*c)>0) and (a>0):
        print('Has local minima')
    if ((a*d)-(b*c)>0) and (a<0):
        print('Has local maxima')
    if ((a*d)-(b*c)<0):
        print('saddle point')
    if ((a*d)-(b*c)==0):
        print('hessian test fails,go by inspection method')
```

```python
#1>  Quadratic form is z=x^2+y^2 and hessian matrix is([2,0],[0,2])
h=np.array([[2,0],[0,2]])
print(f'The eigen values are :{np.linalg.eigvals(h)}')
z=x**2+y**2
ax=plt.axes(projection='3d')
ax.plot_surface(x,y,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

max_min(2,0,0,2)
```
