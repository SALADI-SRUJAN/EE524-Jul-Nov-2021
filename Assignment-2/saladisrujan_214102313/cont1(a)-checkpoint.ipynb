```python
#importing required libraries
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.special as sc
```

# uniform

```python
def uniform_pdf(a,b):
    x=np.linspace(-10,10)
    y=np.piecewise(x,[(x>=a)&(x<=b)],[lambda x:1/(b-a)])
    return y        
u=uniform_pdf(2,5)
plt.plot(x,u,label='u')
v=uniform_pdf(5,9)
plt.plot(x,v,label='v')
w=uniform_pdf(3,8)
plt.plot(x,w,label='w')
plt.title('pdf of uniform distribution')
plt.xlabel('x')
plt.ylabel('pdf of x')
plt.legend()
plt.show()
```

```python
##plotting mean,median and mode
a=2
b=5
u=uniform_pdf(a,b)
plt.plot(x,u,label='u')
mean=(a+b)/2
mode=(b-a)**2/12
plt.plot(mean,0,"s",label='mean,median')
plt.plot(mode,0,"o",label='mode')
plt.legend()
plt.show()
```

```python
#plotting cdf
def uniform_cdf(a,b):
    x=np.linspace(-10,10)
    z=np.piecewise(x,[((x>=a) & (x<=b)),(x>b)],[lambda x:((x-a)/(b-a)),1]) #cdf of uniform x
    return z
u=uniform_cdf(2,5)
plt.plot(x,u,label='u')
v=uniform_cdf(5,9)
plt.plot(x,v,label='v')
w=uniform_cdf(3,8)
plt.plot(x,w,label='w')
plt.title('cdf of uniform distribution')
plt.xlabel('x')
plt.ylabel('cdf of x')
plt.legend()
plt.show()
```

# exponential

```python
#pdf of eponential distribution
def expo(a):
    x=np.linspace(-10,10)
    y=np.piecewise(x,[(x>=0)&(x<=10)],[lambda x:a*2.718**(-a*x)])
    return y
x=np.linspace(-10,10)
u=expo(5)
plt.plot(x,u,label='u')
v=expo(1)
plt.plot(x,v,label='v')
w=expo(3)
plt.plot(x,w,label='w')
plt.xlabel('x')
plt.ylabel('pdf of x')
plt.title('exponential distribution')
plt.legend()
plt.show()
```

```python
#mean,median and variance
a=5
u=expo(5)
plt.plot(x,u,label='u')
mean=1/a
median=0.6931/a
var=a**(-2)
plt.plot(mean,0,"s",label='mean')
plt.plot(median,0,"o",label='median')
print(f"the variance is : {var}")
plt.legend()
plt.show()
```

```python
def expo(a):
    x=np.linspace(-10,10)
    y=np.piecewise(x,[(x>=0)&(x<=10)],[lambda x:1-(2.718)**(-a*x)])
    return y
u=expo(0.5)
plt.plot(x,u,label='u')
v=expo(1)
plt.plot(x,v,label='v')
w=expo(3)
plt.plot(x,w,label='w')
plt.xlabel('x')
plt.ylabel('cdf of x')
plt.title('cdf of exponential distribution')
plt.legend()
plt.show()
```

# gaussian


```python
def gauss(a,b): #a=mu,b=sigma
    x=np.linspace(-10,10)
    y=np.piecewise(x,[(x>=-10)&(x<=10)],[lambda x:(1/b*math.sqrt(2*3.14))*(2.718**(-0.5*((x-a)/b)**2))])
    return y
u=gauss(0,5)
plt.plot(x,u,label='u')
v=gauss(1,5)
plt.plot(x,v,label='v')
w=gauss(2,6)
plt.plot(x,w,label='w')
plt.xlabel('x')
plt.ylabel('pdf of x')
plt.title('pdf of gaussian distribution')
plt.legend()
plt.show()
               
               
```

```python
a=0
b=5
u=gauss(0,5)
plt.plot(x,u,label='u')
mean=a
var=b**2
plt.plot(mean,0,"s",label='mean,median,mode')
print(f"the variance is : {var}")
plt.legend()
plt.show()
```

```python
##cdf of gaussian
from scipy.stats import norm
def gauss_cdf(a,b): #a=mu,b=sigma
    x=np.linspace(-10,10)
    y=norm.cdf(x,loc=a,scale=b)
    return y
u=gauss_cdf(0,0.44721)
plt.plot(x,u,label='u')
v=gauss_cdf(1,5)
plt.plot(x,v,label='v')
w=gauss_cdf(2,6)
plt.plot(x,w,label='w')
plt.xlabel('x')
plt.ylabel('cdf of x')
plt.title('cdf of gaussian distribution')
plt.legend()
plt.show()
```

# gamma

```python
def gamma_pdf(a,b):#a=shape parameter,b=scale parameter
    y=np.piecewise(x,[(x>=0)&(x<=10)],[lambda x:(1/(math.gamma(a)*b**a))*(x**(a-1))*(2.718**(-x/b))])
    return y
x=np.linspace(0,10)    
u=gamma_pdf(1,2)
plt.plot(x,u,label='u')
v=gamma_pdf(5,1)
plt.plot(x,v,label='v')
w=gamma_pdf(9,0.5)
plt.plot(x,w,label='w')
plt.title('pdf of gamma distribution')
plt.ylabel('pdf of x')
plt.legend()
plt.xlabel('x')
plt.show()
```

```python
#mean,mode and variance of gamma distribution
a=5
b=1
x=np.linspace(0,10)
v=gamma_pdf(a,b)
plt.plot(x,v,label='v')
plt.title('gamma distribution')
mean=a*b
mode=(a-1)*b
var=a*b**2
plt.plot(mean,0,"s",label='mean')
plt.plot(mode,0,"o",label='mode')
print(f"the variance is : {var}")
plt.legend()
plt.show()
```

```python
##cdf of gamma distribution
from scipy.stats import gamma
def gamma_cdf(a,b):
    y=gamma.cdf(x,a,loc=0,scale=b)
    return y
x=np.linspace(0,10)
u=gamma_cdf(1,2)
plt.plot(x,u,label='u')
v=gamma_cdf(5,1)
plt.plot(x,v,label='v')
w=gamma_cdf(9,0.5)
plt.plot(x,w,label='w')
plt.title('cdf of gamma distribution')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()
plt.show()
```

# beta distribution

```python
#beta distribution pdf
from scipy.stats import beta
def beta_pdf(a,b):   #a,b=shape parameter
    y=beta.pdf(x,a,b)
    return y
x=np.linspace(0,1)
u=beta_pdf(0.5,0.5)
plt.plot(x,u,label='u')
v=beta_pdf(5,1)
plt.plot(x,v,label='v')
w=beta_pdf(2,5)
plt.plot(x,w,label='w')
plt.title('pdf of beta distribution')
plt.xlabel('x')
plt.ylabel('pdf of x')
plt.legend()
plt.show()
    
```

```python
#mean,median,mode and variance of beta distribution
a=5
b=1
v=beta_pdf(a,b)
x=np.linspace(0,1)
plt.plot(x,v,label='v')
mean=a/(a+b)
median=(a-0.333)/(a+b-0.666) #defined for a,b>1
mode=(a-1)/(a+b-2)           #defined for a,b>1
var=(a*b)/(a+b)**2*(a+b+1)
plt.plot(mean,0,"s",label='mean')
plt.plot(median,0,"o",label='median')
plt.plot(mode,0,"d",label='mode')
print(f'the variance is :{var}')
plt.title('beta distribution')
plt.legend()
plt.show()
```

```python
##cdf of beta
import scipy.special as sc
def beta_cdf(a,b):
    y=np.piecewise(x,[(x>=0)&(x<=1)],[lambda x:sc.betainc(a,b,x)])
    return y
x=np.linspace(0,1)
u=beta_cdf(2,2)
plt.plot(x,u,label='u')
v=beta_cdf(5,1)
plt.plot(x,v,label='v')
w=beta_cdf(1,3)
plt.plot(x,w,label='w')
plt.title('cdf of beta distribution')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()
plt.show()
    
```

# chi square distribution

```python
def chisquare_pdf(k): #k=degrees of freedom
    x=np.linspace(-10,10)
    y=np.piecewise(x,[(x>=0)&(x<=10)],[lambda x:(x**((k/2)-1)*(2.718**(-x/2)))/(2**(k/2)*math.gamma(k/2))])#pdf formula of chi square
    return y
x=np.linspace(-10,10)
u=chisquare_pdf(4)
plt.plot(x,u,label='u')
v=chisquare_pdf(6)
plt.plot(x,v,label='v')
w=chisquare_pdf(3)
plt.plot(x,w,label='w')
plt.title('pdf of chi-suare distribution')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.legend()
plt.show()


```

```python
#mean median and mode,variance of chisquare
k=6
v=chisquare_pdf(6)
x=np.linspace(-10,10)
plt.plot(x,v,label='v')
mean=k
median=k*(1-(2/9*k))**3
mode=max(k-2,0)
var=2*k
plt.plot(mean,0,"s",label='mean')
plt.plot(median,0,"o",label='median')
plt.plot(mode,0,"d",label='mode')
print(f'the variance is :{var}')
plt.title('chi-square distribution')
plt.legend()
plt.show()
```

```python
##cdf of chi-square
def chisquare_cdf(k):
    x=np.linspace(-10,10)
    y=np.piecewise(x,[(x>=0)&(x<=10)],[lambda x: sc.gammainc(k/2,x/2)/math.gamma(k/2)]) #using sciplot functions for cdf
    return y
u=chisquare_cdf(4)
plt.plot(x,u,label='u')
v=chisquare_cdf(6)
plt.plot(x,v,label='v')
w=chisquare_cdf(3)
plt.plot(x,w,label='w')
plt.title('cdf of chisquare')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()
plt.show()
```
